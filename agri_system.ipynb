```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Simulate sensor data
np.random.seed(42)
n_samples = 500

data = pd.DataFrame({
    'soil_moisture': np.random.uniform(20, 80, n_samples),
    'temperature': np.random.uniform(15, 35, n_samples),
    'humidity': np.random.uniform(40, 90, n_samples),
    'sunlight': np.random.uniform(200, 1200, n_samples),
    'soil_ph': np.random.uniform(5.5, 7.5, n_samples),
})

# Simulate crop yield based on features
data['crop_yield'] = (
    0.3 * data['soil_moisture'] +
    0.2 * data['temperature'] +
    0.1 * data['humidity'] +
    0.25 * data['sunlight'] / 100 +
    0.15 * data['soil_ph']
) + np.random.normal(0, 3, n_samples)

# Train AI model
X = data.drop('crop_yield', axis=1)
y = data['crop_yield']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

# Evaluate and visualize results
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"ðŸŒ± Root Mean Squared Error: {rmse:.2f}")

plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred, color='green')
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
plt.title('Crop Yield Prediction')
plt.grid(True)
plt.show()

# Summary
print("\n=== KEY SENSORS ===")
print(f"âœ“ Temperature Sensor")
print(f"âœ“ Soil Moisture Sensor")
print(f"âœ“ Light Sensor (LDR)")
print(f"âœ“ Humidity Sensor")
print(f"âœ“ Nutrient Sensor")
print(f"âœ“ pH Sensor")
```
